```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('diabetes.csv')

# Define function to plot pie charts for diabetic vs non-diabetic individuals for given symptom
def plot_pie_chart(column_name):
    # Calculate the distribution of diabetes based on the presence/absence of the symptom
    diabetic_counts = data.groupby([column_name, 'Diabetic']).size().unstack().fillna(0)
    
    # Labels for each category (0 = No, 1 = Yes)
    labels = ['Non-Diabetic', 'Diabetic']
    
    # Plot pie charts for symptom presence (1) and absence (0)
    fig, axes = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'aspect': 'equal'})
    fig.suptitle(f"Diabetes Distribution by '{column_name.capitalize()}'", fontsize=16)

    for i, presence in enumerate([0, 1]):
        sizes = diabetic_counts.loc[presence]
        axes[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
        axes[i].set_title(f"{column_name.capitalize()} = {presence} (Absent)" if presence == 0 else f"{column_name.capitalize()} = {presence} (Present)")

    plt.show()

# Generate pie charts for the most correlated symptoms with diabetes
for symptom in ['polyuria', 'polydipsia', 'sudden_weight_loss', 'partial_paresis', 'polyphagia','obesity','weakness','irritability','visual_blurring','muscle_stiffness','delayed_healing','alopecia','itching',]:
    plot_pie_chart(symptom)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('diabetes.csv')

# Filter diabetic individuals with all three symptoms present
diabetic_with_symptoms = data[(data['Diabetic'] == 1) & 
                              (data['polyuria'] == 1) & 
                              (data['polydipsia'] == 1) & 
                              (data['sudden_weight_loss'] == 1)]

# Calculate counts for diabetic individuals
total_diabetic_count = data[data['Diabetic'] == 1].shape[0]
diabetic_with_symptoms_count = diabetic_with_symptoms.shape[0]

# Filter non-diabetic individuals with all three symptoms present
non_diabetic_with_symptoms = data[(data['Diabetic'] == 0) & 
                                  (data['polyuria'] == 1) & 
                                  (data['polydipsia'] == 1) & 
                                  (data['sudden_weight_loss'] == 1)]

# Calculate counts for non-diabetic individuals
total_non_diabetic_count = data[data['Diabetic'] == 0].shape[0]
non_diabetic_with_symptoms_count = non_diabetic_with_symptoms.shape[0]

# Prepare data for pie charts
sizes_diabetic = [diabetic_with_symptoms_count, total_diabetic_count - diabetic_with_symptoms_count]
sizes_non_diabetic = [non_diabetic_with_symptoms_count, total_non_diabetic_count - non_diabetic_with_symptoms_count]
labels = ['With Symptoms (Polyuria, Polydipsia, Sudden Weight Loss)', 'Without All Symptoms']
colors = ['lightcoral', 'lightskyblue']

# Plot pie chart for diabetic individuals
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(sizes_diabetic, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Diabetic Individuals with Polyuria, Polydipsia, and Sudden Weight Loss")

# Plot pie chart for non-diabetic individuals
plt.subplot(1, 2, 2)
plt.pie(sizes_non_diabetic, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Non-Diabetic Individuals with Polyuria, Polydipsia, and Sudden Weight Loss")

plt.show()
```
