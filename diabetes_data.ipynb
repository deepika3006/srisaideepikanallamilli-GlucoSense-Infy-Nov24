Visualizing the percantage of diabetic and non-diabetic people having the symptoms

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('diabetes.csv')

# Define function to plot pie charts for diabetic vs non-diabetic individuals for given symptom
def plot_pie_chart(column_name):
    # Calculate the distribution of diabetes based on the presence/absence of the symptom
    diabetic_counts = data.groupby([column_name, 'Diabetic']).size().unstack().fillna(0)
    
    # Labels for each category (0 = No, 1 = Yes)
    labels = ['Non-Diabetic', 'Diabetic']
    
    # Plot pie charts for symptom presence (1) and absence (0)
    fig, axes = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'aspect': 'equal'})
    fig.suptitle(f"Diabetes Distribution by '{column_name.capitalize()}'", fontsize=16)

    for i, presence in enumerate([0, 1]):
        sizes = diabetic_counts.loc[presence]
        axes[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
        axes[i].set_title(f"{column_name.capitalize()} = {presence} (Absent)" if presence == 0 else f"{column_name.capitalize()} = {presence} (Present)")

    plt.show()

# Generate pie charts for the most correlated symptoms with diabetes
for symptom in ['polyuria', 'polydipsia', 'sudden_weight_loss', 'partial_paresis', 'polyphagia','obesity','weakness','irritability','visual_blurring','muscle_stiffness','delayed_healing','alopecia','itching',]:
    plot_pie_chart(symptom)
```

Here 0 represents positive and 1 represents negative
Here are the pie charts that represents the sympotms of the diabetic people each graph describes the relationship between them and the percentage of the diabetic people having the sympotms and the percentage of diabetic people doesn't having the symptoms 
92.4% of diabetic people have the symptom polyuria and 24.9% doesn't aving the symptom.
96.6% of patients people have the symptom polydipsia and 33.1% doesn't having the symptom.
86.5% of diabetic people have the symptom sudden_weight_loss and 43.5% doesn't having the symptom.
79.7% of diabetic people have the symptom partial_paresis and 46.3% doesn't having the symptom.
63.3% of diabetic people have the symptom obesity and 60.0% doesn't having the symptom.




Findind the percentage of diabetic people having polyuria,polydipsia and sudden weight loss.

```python

# Filter for diabetic individuals with all three symptoms present
diabetic_with_symptoms = data[(data['Diabetic'] == 1) & 
                              (data['polyuria'] == 1) & 
                              (data['polydipsia'] == 1) & 
                              (data['sudden_weight_loss'] == 1)]

# Calculate counts for pie chart
diabetic_with_symptoms_count = diabetic_with_symptoms.shape[0]
total_diabetic_count = data[data['Diabetic'] == 1].shape[0]

# Values for pie chart
sizes = [diabetic_with_symptoms_count, total_diabetic_count - diabetic_with_symptoms_count]
labels = ['Diabetic with Symptoms (Polyuria, Polydipsia, Sudden Weight Loss)', 'Diabetic without All Symptoms']
colors = ['lightcoral', 'lightskyblue']

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Percentage of Diabetic Individuals with Polyuria, Polydipsia, and Sudden Weight Loss")
plt.show()
```

appox 41% of the diabetic individuals have polyuria,polydipsia,and sudden weight loss all together.This indicates that the presence of these symptoms is a strong indicator of diabetics.
